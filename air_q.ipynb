**Forecasting with FB Prophet Algorithm**

FB PROPHER DOCUMENTATION : https://facebook.github.io/prophet/docs/quick_start.html

```python
#convert#ing the date from DD/MM/YYYY to  YYYY/MM/DD
date_info=pd.to_datetime(air_quality_data['Date'])
```

```python
#the formting has been changed to YYYY/MM/DD
print(date_info)
```

```python
time_info=air_quality_data['Time']
```

```python
print(time_info)
```

```python
#changing the time format
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
#lets try to check the datatype of time_info and date_info
print(type(date_info))
print(type(time_info))
```

**pandas serie**s--> 1row and 1 col
**pandas dataframe**--> table multi dimensional

```python
#combining 2 series to a pandas dataframe
data_time=pd.concat([date_info,time_info],axis=1)
```

```python
data_time.head()
```

```python
data_time.shape
```

```python
#combining date and time
data_time['ds']= data_time['Date'].astype(str)+' '+data_time['Time'].astype(str)
```

```python
data_time.head()
```

```python
data_time.info()
```

We have to convert ds from object datatype to data time format

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(data_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

*This is  what similar to the fb prophet algo --> which has 2 cols of ds and y*

```python
from prophet import Prophet
#This is the model which we will use
```

```python
#training the PROPHET MODEL

model= Prophet()
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=365,freq="H")
future.tail()
```

```python
forecast=model.predict()
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=model.plot(forecast)
```

```python
fig2=model.plot_components(forecast)
```
